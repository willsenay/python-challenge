```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_table = pd.merge(ride_data, city_data, on='city', how='left')

# Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Build the dataframes of each type
df_urban = merge_table.set_index('type').loc['Urban', ['city', 'fare', 'driver_count']]
df_rural = merge_table.set_index('type').loc['Rural', ['city', 'fare', 'driver_count']]
df_suburban = merge_table.set_index('type').loc['Suburban', ['city', 'fare', 'driver_count']]
```

```python
# Collect x and y coordinates of each type as well as the number of rides (size of dots)

# urban
urban_fare_avg = df_urban.groupby('city')['fare'].mean()
urban_ride_count = df_urban.groupby('city')['city'].count()
urban_driver_count = df_urban.reset_index().drop_duplicates(subset='city')\
            .sort_values('city').set_index('city')['driver_count']

# suburban
suburban_fare_avg = df_suburban.groupby('city')['fare'].mean()
suburban_ride_count = df_suburban.groupby('city')['city'].count()
suburban_driver_count = df_suburban.reset_index().drop_duplicates(subset='city')\
            .sort_values('city').set_index('city')['driver_count']

# rural
rural_fare_avg = df_rural.groupby('city')['fare'].mean()
rural_ride_count = df_rural.groupby('city')['city'].count()
rural_driver_count = df_rural.reset_index().drop_duplicates(subset='city')\
            .sort_values('city').set_index('city')['driver_count']
```

```python
# Configure Chart

# urban points
plt.scatter(urban_ride_count, urban_fare_avg, s=10*urban_driver_count, c='lightcoral', edgecolor='black',
           linewidth=1, marker='o', alpha=0.8, label='Urban')

# suburban points
plt.scatter(suburban_ride_count, suburban_fare_avg, s=10*suburban_driver_count, c='lightskyblue', edgecolor='black',
           linewidth=1, marker='o', alpha=0.8, label='Suburban')

# rural points
plt.scatter(rural_ride_count, rural_fare_avg, s=10*rural_driver_count, c='gold', edgecolor='black',
           linewidth=1, marker='o', alpha=0.8, label='Rural')

# legend and handle size configuration
for legend_marker in plt.legend(title='City Types').legendHandles:
    legend_marker._sizes = [30]

# remaining labels, text, and grid
plt.title('Pyber Ride Sharing Data (2016)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.grid()
plt.text(42, 35, 'Note:\nCircle size correlates with driver count per city.')

# save image
plt.savefig('images/Pyber_Bubble_Chart.png')

# show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# individual amounts
urban_fares = df_urban['fare'].sum()
suburban_fares = df_suburban['fare'].sum()
rural_fares = df_rural['fare'].sum()

# total amount
total_fares = urban_fares + suburban_fares + rural_fares

# percentages
urban_fare_pct = urban_fares / total_fares
suburban_fare_pct = suburban_fares / total_fares
rural_fare_pct = rural_fares / total_fares

# chart details
percentages = [urban_fare_pct, rural_fare_pct, suburban_fare_pct]
explode = [.1, 0, 0]
labels = ['Urban', 'Rural', 'Suburban']
colors = ['lightcoral', 'gold', 'lightskyblue']

# Build Pie Chart
plt.pie(percentages, explode=explode, labels=labels, startangle=285, shadow=True, autopct='%1.1f%%', colors=colors)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig('images/Pyber_Fare_Pct_Pie_Chart.png')

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# individual amounts
urban_ride_sum = urban_ride_count.sum()
suburban_ride_sum = suburban_ride_count.sum()
rural_ride_sum = rural_ride_count.sum()

# total amount
total_ride_sum = urban_ride_sum + suburban_ride_sum + rural_ride_sum

# percentages
urban_ride_pct = urban_ride_sum / total_ride_sum
suburban_ride_pct = suburban_ride_sum / total_ride_sum
rural_ride_pct = rural_ride_sum / total_ride_sum

# chart details
percentages = [urban_ride_pct, rural_ride_pct, suburban_ride_pct]
colors = ['lightcoral', 'gold', 'lightskyblue']
explode = [.1, 0, 0]
labels = ['Urban', 'Rural', 'Suburban']

# Build Pie Chart

plt.pie(percentages, colors=colors, explode=explode, labels=labels, shadow=True, startangle=263, autopct='%1.1f%%')
plt.title('% of Total Rides by City Type')

# Save Figure
plt.savefig('images/Pyber_Rides_Pct_Pie_Chart.png')

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# individual amounts
urban_driver_sum = urban_driver_count.sum()
suburban_driver_sum = suburban_driver_count.sum()
rural_driver_sum = rural_driver_count.sum()

# total amount
total_driver_sum = urban_driver_sum + suburban_driver_sum + rural_driver_sum

# percentages
urban_driver_pct = urban_driver_sum / total_driver_sum
suburban_driver_pct = suburban_driver_sum / total_driver_sum
rural_driver_pct = rural_driver_sum / total_driver_sum

# chart details
percentages = [urban_driver_pct, rural_driver_pct, suburban_driver_pct]
colors = ['lightcoral', 'gold', 'lightskyblue']
explode = [.1, 0, 0]
labels = ['Urban', 'Rural', 'Suburban']

# Build Pie Charts
plt.pie(percentages, colors=colors, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=230)
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig('images/Pyber_Driver_Pct_Pie_Chart.png')

# Show Figure
plt.show()
```

# Three Observations

1) There is a strong correlation between fares earned and total rides by city type

2) The farther you are from the city, the fewer number of rides are requested, and the more expensive those rides are.

3) There farther into the city you get, the more drivers there are. The competition could be a reason the prices are lower in the city.
